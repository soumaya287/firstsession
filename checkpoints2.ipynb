```python
#EXERCICE2
def fact(n):
    x=1
    for i in range (1,n+1):
        x=x*i
    return x
print(fact(6))
```

```python
#exercice4
def missing_char(str,x):
    str=str.replace(str[x],"")
    return(str)
print(missing_char('kitten',4))
```

```python
#exercice5
import numpy as np 
def array_to_list(new_array):
    l=[]

    for i in range(new_array.shape[0]):
         l.insert(i,list(new_array[i,:]))
    return(l)
tab=np.array([[0,1],[2,3],[4,5]])
print(array_to_list(tab))
```

```python
#exercice6
import numpy as np
x = np.array([0, 1, 2])
y = np.array([2, 1, 0])
print("\nOriginal array1:")
print(x)
print("\nOriginal array1:")
print(y)
print("\nCovariance matrix of the said arrays:\n",np.cov(x, y))
```

```python
#exercice7
import math
import numpy as np
C = 50
H = 30
formula = lambda D: round(math.sqrt((2*C*D)/H))
a, b, c = map(float, input().split(','))
l=[]
l.append(formula(a))
l.append(formula(c))
l.append(formula(b))
converted_list = [str(element) for element in l]
joined_numbers = ",".join(converted_list)
print(joined_numbers)
```
